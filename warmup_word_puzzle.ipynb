```python
with open('/usr/share/dict/words') as f:
    words = f.read().split('\n')

print(words[:5])
```

```python
import pandas as pd
word_df = pd.DataFrame({'words': words})

word_df
```

```python
# lowercase all the words

word_df_lower = word_df['words'].str.lower()
word_df_lower
```

```python
five_letter_words = word_df_lower[word_df_lower.str.len() == 5]
five_letter_words
```

```python
word_df = pd.DataFrame({'words': five_letter_words,
                       'first_letter': five_letter_words.str[0],
                       'last_letters': five_letter_words.str[1:5]})
word_df
```

```python
word_df['value_of_first'] = word_df.first_letter.apply(lambda n: ord(n) - 96)
word_df
```

```python
word_df['value_of_last'] = word_df.last_letters.apply(lambda a: (ord(a[0])-96) + 
                                                      (ord(a[1])-96) + (ord(a[2])-96)
                                                      + (ord(a[3])-96))
word_df
```

```python
answer = word_df[word_df.value_of_first == word_df.value_of_last]
answer
```
