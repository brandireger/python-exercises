```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data

import seaborn as sns

iris = sns.load_dataset('iris')

iris.dtypes
```

## What does the distribution of petal lengths look like?

```python
sns.distplot(iris.petal_length)
```

## Is there a correlation between petal length and petal width?

```python
iris_data = pd.crosstab(iris.petal_length, iris.petal_width)
plt.rc('figure', figsize=(14, 10))

sns.heatmap(iris_data.corr(), center=0, cmap='coolwarm')
```

## Would it be reasonable to predict species based on sepal width and sepal length?

```python
sns.lmplot(x='sepal_width', y='sepal_length', hue='species', data=iris)
```

## Which features would be best used to predict species?

```python
sns.pairplot(iris, hue='species')
```

```python
sns.lmplot(y='petal_width', x='sepal_length', hue='species', data=iris)
```

## Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
def get_db_url(db_name):
    from env import host, user, password
    url = f'mysql+pymysql://{user}:{password}@{host}/{db_name}'
    return url
```

```python
url = get_db_url('chipotle')

chipotle = pd.read_sql('SELECT * FROM orders', url)
chipotle
```

```python
chipotle['item_price'] = chipotle['item_price'].replace('[\$]', '', regex=True).astype(float)

bar_chart = chipotle.groupby('item_name').item_price.agg(['sum']).nlargest(n=4, columns='sum')
type(bar_chart)
```

```python
bar_chart
```

```python
sns.barplot(data=bar_chart, y='sum', x=bar_chart.index)
```
