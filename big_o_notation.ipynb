```python
import matplotlib.pyplot as plt
import math
```

```python
x = range(1,15)
y1 = [0*x + 1 for x in x]
y2 = [math.log(x) for x in x]
y3 = [x for x in x]
y4 = [x * math.log(x) for x in x]
y5 = [x**2 for x in x] 
y6 = [2**x for x in x]
y7 = [math.factorial(x) for x in x] 
y8 = [x**x for x in x]

plt.figure(figsize=(14, 14))

plt.plot(y1, label="O(1)")
plt.plot(y2, label="O(log n)")
plt.plot(y3, label="O(n)")
plt.plot(y4, label="O(n log n)")
plt.plot(y5, label="O(n^2)")
plt.plot(y6, label="O(2^n)")
plt.plot(y7, label="O(n!)")
plt.plot(y8, label="O(n^n)")

plt.legend()
plt.title("Big O Notation")
plt.xlabel("Elements")
plt.ylabel("Operations")
plt.ylim(0,15)
plt.xlim(0,10)

plt.show()
print(plt)
```
